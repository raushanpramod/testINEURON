```python
# All numbers that are a factor of 7 but not a factor of 5 b/w 2000 to 3200
```

```python
div_7 = []

for x in range(2000, 3201):
    if x % 7 == 0 and x % 35 != 0:
        div_7.append(x)
print(div_7)
```

```python
# Write a Python program to find the volume of a sphere with diameter 12 cm.
```

```python
import math
r = 12
v = 4/3 * math.pi * r**3
print('volume = ', round(v, 2),"cm^3")
```
